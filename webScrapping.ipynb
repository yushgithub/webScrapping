```python
import pandas as pd

# Initialize empty lists for data
money_list = []
menu_list = []
link_list = []
title_list =[]




# Fetch spans for money (Pay and Donation)
money_spans = soup.find('div', class_='category-item null rounded-2xl cursor-pointer theme-dark-blue overflow-hidden relative box-border inline-block mr-3.5 rtl:ml-3.5 rtl:mr-0 rtl:last:ml-0 last:mr-0').find_all('span')
for span in money_spans:
    money_list.append(span.get_text(strip=True))

# Fetch spans for menu (Food and DineOut)
menu_spans = soup.find('div', class_='category-item null rounded-2xl cursor-pointer theme-green overflow-hidden relative box-border inline-block mr-3.5 rtl:ml-3.5 rtl:mr-0 rtl:last:ml-0 last:mr-0').find_all('span')
for span in menu_spans:
    menu_list.append(span.get_text(strip=True))

# Fetch the title link
title = soup.find('a', class_="header-nav-link relative px-5 py-4 z-10 lg:px-0 lg:py-6 nuxt-link-exact-active nav-active block").get_text(strip=True)
title_list.extend([title] * max(len(money_list), len(menu_list),len(link_list)))

# Fetch the image link
link = soup.find('img', class_="inline mb-1").get('src')
link_list.extend([link] * max(len(money_list), len(menu_list)))

# Determine the maximum length of the lists to iterate over
max_len = max(len(money_list), len(menu_list))

# Create a DataFrame from the lists
df = pd.DataFrame({'Title': title_list, 'Money': money_list, 'Menu': menu_list, 'Link': link_list})

# Display the DataFrame
display(df)
```
